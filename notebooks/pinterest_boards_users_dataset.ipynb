```python
%load_ext autoreload
%autoreload 2

from pprint import pprint
import pandas as pd
import pymongo
import numpy as np
import time
import networkx as nx
import operator
import vislab.util
```

```python
vislab.util.print_collection_counts()
db = vislab.util.get_mongodb_client()['pinscraping']
```

## Load boards collection into a DataFrame.

```python
pprint(db['boards'].find_one())
```

```python
t = time.time()
df = pd.DataFrame(
    list(db['boards'].find(fields={
        'thumb_imgs': False, 'cover_img': False, 'board_url': False, 'source': False
    }))
)
print('Assembling boards took {:.2f} s'.format(time.time() - t))
N = df.shape[0]

# Drop duplicates
boards_df = df.drop_duplicates(cols=['username', 'board_name'], take_last=True)
print("Dropped {} duplicates, now have {} rows".format(N - boards_df.shape[0], boards_df.shape[0]))
```

## Load users collection as a DataFrame

```python
print db['users'].count()
user_document = db['users'].find({'username': u'pixelmao'})[0]
pprint(user_document)
```

```python
# Construct DataFrame of users and make sure that only unique
# queries are stored in the 'query' column.
users_df = pd.DataFrame(list(db['users'].find(fields={'_id': False})))
users_df['query'] = [np.unique(_) for _ in users_df['query']]
```

```python
# Join the users and boards DataFrames on username column.
# Then expand the 'query' column to separate boolean columns,
# one for each possible query.
boards_users_df = pd.merge(users_df, boards_df, on='username')
print boards_users_df.shape
print boards_users_df.columns
```

```python
boards_users_df_expanded = pd.DataFrame(
    data=[
        dict([('query_' + _, True) for _ in row])
        for row in boards_users_df['query']
    ],
    index=boards_users_df.index
).fillna(False)
print boards_users_df_expanded.shape
print boards_users_df_expanded.columns
```

```python
boards_users_df = boards_users_df.join(boards_users_df_expanded)
boards_users_df.set_index('_id', inplace=True)
del boards_users_df['query']
boards_users_df
```
