### Date distribution plot.

```python
import vislab.gg

def plot_date_distribution(df):
    dates = df['date'].dropna().apply(lambda x: int(re.search('(\d+)', x).groups()[0]))
    
    filtered_dates = dates[(dates > 1400) & (dates < 2015)]
    fig = plt.figure(figsize=(12,4))
    ax = fig.add_subplot(111)
    ax.hist(filtered_dates, bins=50)
    ax.set_xlim([1400, 2020])
    ax.set_title('Date Frequency')
    ax.set_xlabel('Year')
    vislab.gg.rstyle(ax)
    return fig
    
fig = plot_date_distribution(df)
fig.savefig('/Users/sergeyk/work/aphrodite-writeup/figures/wikipaintings_date.pdf', bbox_inches='tight')
```
